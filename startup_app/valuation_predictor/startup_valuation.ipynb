```python
def grow(x):
    if x == 'FinTech':  # BCG
        return 14
    elif x == 'TravelTech':  # Research & Market
        return 6.6
    elif x == 'AgriTech':  # Global Research Company
        return 12.9
    elif x == 'HealthTech':  # Precedence
        return 11.68
    elif x == 'FoodTech':
        return 8.2
    elif x == 'EdTech':
        return 14
    elif x == 'Retail':
        return 2.2
    else:
        return None  # If no match, return None

# Apply function to 'Sector' column
df_new_trend['growth_rate_new'] = df_new_trend['Sector'].apply(grow)
```

```python
##SINCE THE DATASET IS NOT PLOTTABLE WILL DO LITTLE PREPROCESSING ---> CONVERTING M & B INTO NOs and removing nulls

def millions_to_no(x):
    if isinstance(x, str): ##ininstance() is used to check if object belong to specific datatype
        x = x.replace('$','').strip().upper()
        x = x.replace('£','£ ')
        #* 0.73
        x = x.replace('CA','CA ')
        if x.endswith('M') and x.startswith('£'):
            x = x[1:] #remove first character
            y = float(x[:-1]) * 1.08 * 1_000_000 ##remove last character then applying multiplication
            return y
        elif x.endswith('M') and x.startswith('CA'):
            x = x.replace('CA','').strip()
            y = float(x[:-1]) * 0.73 * 1_000_000
            return y
        elif x.endswith('M'):
            return float(x[:-1]) * 1_000_000
        elif x.endswith('B'):
            return float(x[:-1]) * 1_000_000_000
        elif x.endswith('K'):
            return float(x[:-1]) * 1_000
        else:
            return x
    return x  # Already numeric or NaN

resultant['Amount_Raised'] = resultant['Amount_Raised'].apply(millions_to_no)
```

1. **Capital Efficiency**

   * 🔹 **Low ratio** → Startup raised **less money** compared to its valuation → Possibly efficient or lean.
   * 🔹 **High ratio** → Raised a **lot relative to valuation** → Could indicate capital-intensive business or overfunding.

2. **Investor Sentiment**

   * High ratio might suggest investors are putting in a lot **but valuation hasn't caught up** → Less confidence or early stage.
   * Low ratio with high valuation → Investors see strong potential or hype.

3. **Stage Indicator**

   * Early-stage startups often have **high ratios** (low valuations but need cash).
   * Mature startups may have **low ratios** (high valuation, less funding needed).

4. **Red Flags or Outliers**

   * Very **high ratios** could flag startups that raised too much without growth.
   * Very **low ratios** could be **overvalued** companies with little backing.
